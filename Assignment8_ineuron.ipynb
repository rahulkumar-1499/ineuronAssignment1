In Local Scope the variables are created within a function and can be accessed within the function.They are destroyed once the function executes completely.
For ex:-

def sumval(a,b):
    sum=a+b
    print(sum)     //the value will be printed

In global scope the variables are defined outside any function and can be accessed and modified at anytime and from anywhere.

for eg:-

6.In Python, you can use the return statement to return multiple values from a function by separating them with commas. When a function encounters a return statement with multiple values, it packs them into a tuple and returns the tuple. 

for eg:-

def multiple_values():
    value1 = 10
    value2 = "Hello, World!"
    value3 = [1, 2, 3]

    
result1, result2, result3 = multiple_values()


print(result1)  # Output: 10
print(result2)  # Output: Hello, World!
print(result3)  # Output: [1, 2, 3]


7.Pass by value is the actual value passed to the function and the function makes another copy of the argument which is passed whiling calling the function.If we make any chances to the varible which is passed as an argument to the function then there is no effect to the original value.

For eg:-

    x = x + 10
    print("Inside function:", x)

Pass by reference the the address of the actual varible which is passed as a parameter to the function.Hence if we try to change value to the referenced variable then the actual variable wll be affected.

For eg:-

    lst.append(4)
    print("Inside function:", lst)

```python
8.
import math

def perform_operations(x):
    log_result = math.log(x)

    exp_result = math.exp(x)

  
    power_result = math.pow(2, x)

  
    sqrt_result = math.sqrt(x)

  
    results = {
        'logarithmic_function': log_result,
        'exponential_function': exp_result,
        'power_function_base_2': power_result,
        'square_root': sqrt_result
    }

    return results


input_value = 4.0
results = perform_operations(input_value)


print(f"Logarithmic function (log x): {results['logarithmic_function']}")
print(f"Exponential function (exp(x)): {results['exponential_function']}")
print(f"Power function with base 2 (2^x): {results['power_function_base_2']}")
print(f"Square root: {results['square_root']}")
```
